## Insight (Answering the Question)

11.	Is there any relationship between who a person works for (their manager) and their performance score? <br>**(10 points)**<br>
12.	What are our best recruiting sources if we want to ensure a low ratio of termination? <br>**(10 points)**

```python
# Code Number 11 for Dataset Row Table Relationship Analysis

# Import tabulate for formatted table display
from tabulate import tabulate

url = "https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/HRDataset_v14.csv"
df = pd.read_csv(url)

performanceMapping = {
    "Exceeds": 4,
    "Fully Meets": 3,
    "Needs Improvement": 2,
    "PIP": 1
}

df['PerformanceScore_Numeric'] = df['PerformanceScore'].map(performanceMapping)

avg_perform_by_manager = df.groupby('ManagerName').agg(
    AveragePerformanceScore=('PerformanceScore_Numeric', 'mean'),
    EmployeeCount=('Employee_Name', 'count')  # Counting the number of employees
).reset_index()

top_10_managers = avg_perform_by_manager.nlargest(10, 'AveragePerformanceScore')

# Display the top 10 managers along with the total number of employees they manage
print("Top 10 Managers by Average Performance Score and Number of Employees:")
print(tabulate(top_10_managers[['ManagerName', 'AveragePerformanceScore', 'EmployeeCount']],
               headers='keys', tablefmt='grid', showindex=False))

# Memvisualisasikan rata-rata performance scores dari manager
plt.figure(figsize=(12, 8))
sns.barplot(x='AveragePerformanceScore', y='ManagerName', data=avg_perform_by_manager, palette='viridis', hue='ManagerName', dodge=False)
plt.title('Average Performance Scores by Manager')
plt.xlabel('Average Performance Score')
plt.ylabel('Manager Name')
# plt.axvline(x=top_10_managers['AveragePerformanceScore'].mean(), color='red', linestyle='--', label='Mean of Top 10')
plt.legend()
plt.show()

# Import Optional Statistical Analysis (ANOVA) Library
from scipy import stats

# Mendapatkan  performance scores dari masing-masing  manager
performance_scores = [group['PerformanceScore_Numeric'].values for name, group in df.groupby('ManagerName')]

# Mengoperasikan ANOVA
anova_result = stats.f_oneway(*performance_scores)
print("\n================================")
print(f"ANOVA Analysis Result: \nF-statistic = {anova_result.statistic} \np-value = {anova_result.pvalue}")
print("================================")
```

### Hasil dan Penjelasan No. 11

Hasil dari **F-statistic** dan **p-value** yang diperoleh dalam analisis ANOVA di atas adalah sebagai berikut:

- F-statistic = 0.5519835874490843
- p-value = 0.9417523898104365

"**F-statistic**" merupakan ukuran dari rasio variabilitas antara kelompok (dalam hal ini, kelompok manajer) terhadap variabilitas di dalam kelompok (variasi di antara karyawan yang bekerja untuk manajer yang sama). Dimana, semakin tinggi nilai **F-statistic**, maka akan semakin besar perbedaan antara kelompok yang dibandingkan dengan perbedaan di dalam kelompok.

Nilai **F-statistic** yang sangat kecil seperti pada hasil yang telah didapatkan, yakni "**0.55**", menunjukkan bahwa variabilitas di dalam setiap kelompok manajer hampir sama dengan variabilitas antara kelompok manajer. Hal ini menunjukkan bahwa perbedaan performa antara manajer tidak terlalu besar atau signifikan.

---

"**p-value**" menunjukkan probabilitas dimana perbedaan yang diamati terjadi secara kebetulan jika tidak ada hubungan yang sebenarnya (yaitu jika *null hypothesis* benar). Dimana, "**Null Hypothesis (H0)**" merupakan suatu pengujian dengan hasil atau resultnya akan meanandakan bahwa '*tidak ada perbedaan yang signifikan antara rata-rata performa karyawan berdasarkan manajernya.*'.

Maka dari itu, jika **p-value** = "**0.9417**". Hal ini menandakan bahwasanya hasil dari analisis ANOVA akan sangat besar atau jauh lebih tinggi dari ambang signifikan yang umum digunakan, yaitu "**0.05**". Dimana, berarti akan ada probabilitas sebesar "**94.17%**" yang menunjukkan bahwa perbedaan yang terlihat di skor performa antara karyawan di bawah manajer yang berbeda terjadi secara kebetulan saja. Dikarenakan **p-value** jauh lebih besar daripada ambang "**0.05**", dapat dikatakan kami tidak memiliki bukti yang cukup untuk menolak '**null hypothesis**'. Dengan kata lain, tidak ada perbedaan signifikan antara skor performa berdasarkan siapa manajer dari karyawan tersebut. Ditambah lagi, berdasarkan hasil **F-statistic** sebesar "**0.55**" yang akan mendukung hasil dari penganalisaan ini, dimana akan menunjukkan bahwa variasi performa di antara manajer adalah ***kecil***, dan sebagian besar variabilitas dalam performa berasal dari faktor lain selain siapa manajernya.

---

Dengan demikian, dapat disimpulkan hasil dari analisis dengan menggunakan ANOVA ini menunjukkan "**tidak ada hubungan yang signifikan**" antara siapa manajer karyawan dengan performa karyawan. Perbedaan skor performa yang diamati kemungkinan besar terjadi secara kebetulan dan bukan karena pengaruh dari manajernya.

```python
# Code Number 12 for Termination Ratio Best Recruiting Sources Analysis

termination_status = data.groupby(['RecruitmentSource', 'EmpStatusID']).size().unstack(fill_value = 0)

# hitung termination rate untuk setiap source recruitment
# EmpStatusID = 1 = active employee (masih bekerja)
# Hitung total karyawan dan karyawan yang diberhentikan per sumber rekrutmen
termination_status = data.groupby('RecruitmentSource').agg(
    TotalEmployees=('Employee_Name', 'count'),  # Menghitung total karyawan
    TerminatedEmployees=('Termd', 'sum')  # Menghitung jumlah karyawan yang diberhentikan (Termd = 1)
).reset_index()

# Hitung rasio terminasi (TerminationRate)
termination_status['TerminationRatio'] = termination_status['TerminatedEmployees'] / termination_status['TotalEmployees']

# Urutkan berdasarkan rasio terminasi terendah
status_sorted = termination_status.sort_values(by = 'TerminationRatio', ascending = True)

# Display the sorted termination rates with total and terminated employees using tabulate
print("\nTermination Ratio by Recruitment Source with Total and Terminated Employees:")
print(tabulate(status_sorted, headers='keys', tablefmt='grid', showindex=False))

# Memvisualisasikan Hasil dengan Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=termination_status, x='RecruitmentSource', y='TerminationRatio', palette='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.title('Termination Ratio by Recruitment Source')
plt.xlabel('Recruitment Source')
plt.ylabel('Termination Ratio')
plt.show()
```
